```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker

def prettyshow(grads, legend, save='image.pdf', ylabel='Gradient Norm'):
    plt.rcParams['figure.figsize'] = 12, 8
    
    plt.figure()
    xlen = len(grads[0])
    colors = ['green', 'red', 'blue', 'orange', 'purple', 'brown']
    markers = ['d', '^', 'o', '<', '*', 's']
    idx_set = np.arange(0, xlen, xlen//10)
    
    for i in range(len(grads)):
        plt.semilogy(0, grads[i][0], color=colors[i], marker=markers[i], markersize=8)
    
    for i in range(len(grads)):
        for idx in idx_set:
            plt.semilogy(idx, grads[i][idx], color=colors[i], marker=markers[i], markersize=15, linestyle='None')
    
    for i in range(len(grads)):
        plt.semilogy(np.arange(xlen), grads[i], linewidth=4, color=colors[i])
    
    # 23
    plt.legend(legend, prop={'size': 35, 'weight': 'bold'}, loc='upper left', 
               bbox_to_anchor=(0.05, 1), ncol=1)
    
    plt.xlabel('Comm. Rounds', fontsize=47, fontweight='bold')
    plt.ylabel(ylabel, fontsize=47, fontweight='bold')
    
    # 
    ax = plt.gca()
    formatter = ticker.ScalarFormatter(useMathText=True)
    formatter.set_scientific(False)
    formatter.set_powerlimits((3, 6))  # 1e31e6
    ax.xaxis.set_major_formatter(formatter)
    
    # x_tick100
    def format_x_tick(x, pos):
        return f'{x/1:.0f}'
    
    ax.xaxis.set_major_formatter(ticker.FuncFormatter(format_x_tick))
    
    # 
    ax.xaxis.set_major_locator(ticker.MaxNLocator(nbins=5))
    
    # 
    plt.xticks(fontsize=35, fontweight='bold')
    plt.yticks(fontsize=35, fontweight='bold')
    ax.xaxis.get_offset_text().set_fontsize(25)  # 
    ax.xaxis.get_offset_text().set_fontweight('bold')
    plt.title("Grid", fontsize=47, fontweight='bold')
    
    plt.grid(True)
    plt.subplots_adjust(left=0.21, bottom=0.15, right=0.75)  # 
    plt.savefig(save, bbox_inches='tight')
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker

def prettyshow_accuracy(grads, legend, save='image.pdf', ylabel='Gradient Norm'):
    plt.rcParams['figure.figsize'] = 12, 8
    
    plt.figure()
    xlen = len(grads[0])
    colors = ['green', 'red', 'blue', 'orange', 'purple', 'brown']
    markers = ['d', '^', 'o', '<', '*', 's']
    idx_set = np.arange(0, xlen, xlen//10)
    
    for i in range(len(grads)):
        plt.plot(0, grads[i][0], color=colors[i], marker=markers[i], markersize=8)
    
    for i in range(len(grads)):
        for idx in idx_set:
            plt.plot(idx, grads[i][idx], color=colors[i], marker=markers[i], markersize=15, linestyle='None')
    
    for i in range(len(grads)):
        plt.plot(np.arange(xlen), grads[i], linewidth=4, color=colors[i])
    
    plt.legend(legend, prop={'size': 35, 'weight': 'bold'}, loc='upper left', 
        bbox_to_anchor=(0.05, 0.49), ncol=1)
    plt.xlabel('Comm. Rounds', fontsize=47, fontweight='bold')
    plt.ylabel(ylabel, fontsize=47, fontweight='bold')
    
    #  y  0  100
    plt.ylim(0, 100)
    
    #  y  0%, 20%, 40%, 60%, 80%, 100%
    plt.yticks([0, 20, 40, 60, 80, 100], ['0%', '20%', '40%', '60%', '80%', '100%'], fontsize=35, fontweight='bold')
    
    # 
    ax = plt.gca()
    formatter = ticker.ScalarFormatter(useMathText=True)
    formatter.set_scientific(False)
    formatter.set_powerlimits((3, 6))  # 1e31e6
    ax.xaxis.set_major_formatter(formatter)
    
    # 
    ax.xaxis.set_major_locator(ticker.MaxNLocator(nbins=5))
    
    # 
    plt.xticks(fontsize=35, fontweight='bold')
    ax.xaxis.get_offset_text().set_fontsize(25)  # 
    ax.xaxis.get_offset_text().set_fontweight('bold')
    
    plt.grid(True)
    plt.title("Nearest Neighbor", fontsize=47, fontweight='bold')
    plt.subplots_adjust(left=0.21, bottom=0.15)
    plt.subplots_adjust(left=0.21, bottom=0.15, right=0.75)
    plt.savefig(save, bbox_inches='tight')
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker

def new_prettyshow_accuracy(grads, legend, start, end, save='image.pdf', ylabel='Gradient Norm'):
    # Set figure size
    plt.rcParams['figure.figsize'] = 12, 8
    
    # Create figure and main axes
    fig, ax = plt.subplots()
    xlen = len(grads[0])
    colors = ['green', 'red', 'blue', 'orange', 'purple', 'brown']
    markers = ['d', '^', 'o', '<', '*', 's']
    idx_set = np.arange(0, xlen, xlen//10)
    
    # Plot initial points at x=0
    for i in range(len(grads)):
        ax.plot(0, grads[i][0], color=colors[i], marker=markers[i], markersize=8)
    
    # Plot marker points at intervals
    for i in range(len(grads)):
        for idx in idx_set:
            ax.plot(idx, grads[i][idx], color=colors[i], marker=markers[i], 
                    markersize=15, linestyle='None')
    
    # Plot full curves
    for i in range(len(grads)):
        ax.plot(np.arange(xlen), grads[i], linewidth=4, color=colors[i])
    
    # Configure main plot
    ax.legend(legend, prop={'size': 35, 'weight': 'bold'}, loc='upper left', 
              bbox_to_anchor=(0.05, 0.49), ncol=1)
    ax.set_xlabel('Comm. Rounds', fontsize=47, fontweight='bold')
    ax.set_ylabel(ylabel, fontsize=47, fontweight='bold')
    
    ax.set_ylim(0, 100)
    ax.set_yticks([0, 20, 40, 60, 80, 100])
    ax.set_yticklabels(['0%', '20%', '40%', '60%', '80%', '100%'], 
                       fontsize=35, fontweight='bold')
    
    formatter = ticker.ScalarFormatter(useMathText=True)
    formatter.set_scientific(False)
    formatter.set_powerlimits((3, 6))
    ax.xaxis.set_major_formatter(formatter)
    ax.xaxis.set_major_locator(ticker.MaxNLocator(nbins=5))
    ax.tick_params(axis='x', labelsize=35)
    ax.xaxis.get_offset_text().set_fontsize(25)
    ax.xaxis.get_offset_text().set_fontweight('bold')
    
    ax.grid(True)
    ax.set_title("Grid", fontsize=47, fontweight='bold')
    
    # Add inset subplot for (start, end) interval
    inset_left, inset_bottom, inset_width, inset_height = [0.6, 0.6, 0.35, 0.35]  # Position and size
    ax_inset = fig.add_axes([inset_left, inset_bottom, inset_width, inset_height])
    
    for i in range(len(grads)):
        ax_inset.plot(np.arange(start, end), grads[i][start:end], 
                      linewidth=2, color=colors[i])
    
    ax_inset.set_ylim(97.5, 98.3)
    ax_inset.set_yticks([97.5, 97.7, 97.9, 98.1, 98.3])
    ax_inset.set_yticklabels([ '97.5%', '97.7%', '97.9%', '98.1%', '98.3%'], 
                             fontsize=15, fontweight='bold')
    ax_inset.set_xticks(np.arange(start, end+10, (end - start)//5))
    ax_inset.set_xticklabels(np.arange(start, end+10, (end - start)//5), 
                             fontsize=15, fontweight='bold')
    ax_inset.grid(True)
    
    # Adjust layout and save
    plt.subplots_adjust(left=0.1, bottom=0.15, right=1, top=1)
    plt.savefig(save, bbox_inches='tight')
    plt.show()
```
