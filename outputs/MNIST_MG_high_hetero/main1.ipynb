```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = pd.read_csv("./MG=1, EXP16, HIGH HETERO, PullDiag_GT, lr=0.005, n_nodes=16, batch_size=128, 2025-03-27.csv")
df2 = pd.read_csv("./MG=5, EXP16, HIGH HETERO, PullDiag_GT, lr=0.005, n_nodes=16, batch_size=128, 2025-03-27.csv")
df3 = pd.read_csv("./MG=10, EXP16, HIGH HETERO, PullDiag_GT, lr=0.005, n_nodes=16, batch_size=128, 2025-03-27.csv")
```

```python
a = "train_loss(total)"
s = 10000

plt.semilogy(df1[a][:s], label="MG=1")
plt.semilogy(df2[a][:s], label="MG=5")
plt.semilogy(df3[a][:s], label="MG=10")
plt.legend()
plt.show()
```

```python
a = "test_accuracy(average)"
s = 10000

plt.plot(df1[a][:s], label="MG=1")
plt.plot(df2[a][:s], label="MG=5")
plt.plot(df3[a][:s], label="MG=10")
plt.legend()
plt.show()
```
