```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('./MG=2, Ring16, PullDiag_GT, lr=0.005, n_nodes=16, batch_size=128, 2025-03-26.csv', index_col=0)
df2 = pd.read_csv('./MG=4, Ring16, PullDiag_GT, lr=0.005, n_nodes=16, batch_size=128, 2025-03-26.csv', index_col=0)
df3 = pd.read_csv('./MG=8, Ring16, PullDiag_GT, lr=0.005, n_nodes=16, batch_size=128, 2025-03-26.csv', index_col=0)
df4 = pd.read_csv('./MG=1, Ring16, PullDiag_GT, lr=0.005, n_nodes=16, batch_size=128, 2025-03-26.csv', index_col=0)
```

```python
a = "train_loss(total)"
s = 10000

plt.semilogy(df1[a][:s], label="MG=2")
plt.semilogy(df2[a][:s], label="MG=4")
plt.semilogy(df3[a][:s], label="MG=8")
plt.semilogy(df4[a][:s], label="MG=1")
plt.legend()
plt.show()
```

```python
a = "test_accuracy(average)"
s = 10000

plt.plot(df1[a][:s], label="MG=2")
plt.plot(df2[a][:s], label="MG=4")
plt.plot(df3[a][:s], label="MG=8")
plt.plot(df4[a][:s], label="MG=1")
plt.legend()
plt.show()
```
