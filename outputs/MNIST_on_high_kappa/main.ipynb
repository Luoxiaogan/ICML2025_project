```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('MG=1,Xinmeng, PullDiag_GT, lr=0.005, n_nodes=15, batch_size=128, 2025-03-30.csv')
df2 = pd.read_csv('MG=5,Xinmeng, PullDiag_GT, lr=0.005, n_nodes=15, batch_size=128, 2025-03-30.csv')
df3 = pd.read_csv('MG=10,Xinmeng, PullDiag_GT, lr=0.005, n_nodes=15, batch_size=128, 2025-03-30.csv')

a = "train_loss(total)"

plt.semilogy(df1[a], label='MG=1')
plt.semilogy(df2[a], label='MG=5')
plt.semilogy(df3[a], label='MG=10')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Loss vs Iteration')
plt.legend()
plt.grid()
plt.show()
```
