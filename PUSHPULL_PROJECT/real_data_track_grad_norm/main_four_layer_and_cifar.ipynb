```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('./four_layer/grad_norm_Exp_test, PushPull, lr=0.005, n_nodes=4, batch_size=128, 2025-03-23.csv')
df2 = pd.read_csv('./four_layer/grad_norm_Exp_test, PushPull, lr=0.005, n_nodes=8, batch_size=128, 2025-03-23.csv')
df3 = pd.read_csv('./four_layer/grad_norm_Exp_test, PushPull, lr=0.005, n_nodes=16, batch_size=128, 2025-03-23.csv')
df4 = pd.read_csv('./four_layer/grad_norm_Exp_test, PushPull, lr=0.005, n_nodes=32, batch_size=128, 2025-03-23.csv')

a = "avg_grad_norm"
s = 200000
plt.semilogy(df1[a][:s], label="n_nodes=4")
plt.semilogy(df2[a][:s], label="n_nodes=8")
plt.semilogy(df3[a][:s], label="n_nodes=16")
plt.semilogy(df4[a][:s], label="n_nodes=32")

plt.title("4 layer, bs = 128, lr = 0.005")
plt.legend()
plt.show()
```
