```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("./MNIST/Exp_test, PushPull, lr=0.02, n_nodes=2, batch_size=128, 2025-03-14.csv")
df2 = pd.read_csv("./MNIST/Exp_test, PushPull, lr=0.03, n_nodes=2, batch_size=128, 2025-03-14.csv")
df3 = pd.read_csv("./MNIST/Exp_test, PushPull, lr=0.008, n_nodes=2, batch_size=128, 2025-03-14.csv")
df4 = pd.read_csv("./MNIST/Exp_test, PushPull, lr=0.009, n_nodes=2, batch_size=128, 2025-03-14.csv")
```

```python
a = "train_loss(total)"
s = 200

plt.semilogy(df1[a][:s], label="lr=0.02")
plt.semilogy(df2[a][:s], label="lr=0.03")
plt.semilogy(df3[a][:s], label="lr=0.008")
plt.semilogy(df4[a][:s], label="lr=0.009")
plt.title("N=2")
plt.legend()
plt.show()
```

```python
a = "test_accuracy(average)"

s = 200
t = 100

plt.plot(df1[a][t:s], label="lr=0.02")
plt.plot(df2[a][t:s], label="lr=0.03")
plt.plot(df3[a][t:s], label="lr=0.008")
plt.plot(df4[a][t:s], label="lr=0.009")
plt.title("N=2")
plt.legend()
plt.show()
```
