```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("./output/Exp_test, PushPull, lr=0.01, n_nodes=2, batch_size=16, 2025-03-21.csv")
df2 = pd.read_csv("./output/Exp_test, PushPull, lr=0.01, n_nodes=4, batch_size=16, 2025-03-21.csv")
df3 = pd.read_csv("./output/Exp_test, PushPull, lr=0.01, n_nodes=8, batch_size=16, 2025-03-21.csv")
df4 = pd.read_csv("./output/Exp_test, PushPull, lr=0.01, n_nodes=16, batch_size=16, 2025-03-21.csv")
df5 = pd.read_csv("./output/Exp_test, PushPull, lr=0.01, n_nodes=32, batch_size=16, 2025-03-21.csv")
```

```python
plt.semilogy(df1['train_loss_per_batch'], label='n_nodes=2')
plt.semilogy(df2['train_loss_per_batch'], label='n_nodes=4')
plt.semilogy(df3['train_loss_per_batch'], label='n_nodes=8')
plt.semilogy(df4['train_loss_per_batch'], label='n_nodes=16')
plt.semilogy(df5['train_loss_per_batch'], label='n_nodes=32')
plt.ylabel('loss')
plt.xlabel('batch index')
plt.title('Train Loss per batch, EXP graph')
plt.legend()
plt.show()
```
