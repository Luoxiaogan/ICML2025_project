```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# 获取所有 CSV 文件
output_dir = '/Users/luogan/Code/ICML2025_project/PUSHPULL_PROJECT/test/output'
csv_files = [f for f in os.listdir(output_dir) if f.endswith('.csv')]

sigma_n_list = []
variance_y_list = []
variance_pi_B_list = []

for csv_file in csv_files:
    # 提取 sigma_n 值
    sigma_n = float(csv_file.replace('sigma_n_', '').replace('.csv', ''))
    # 读取 CSV 文件
    df = pd.read_csv(os.path.join(output_dir, csv_file))
    # 计算每组数据的均值（示例：取所有时间步的均值）
    avg_variance_y = df['variance_y'].mean()
    avg_variance_pi_B = df['variance_pi_B_ones_T_g'].mean()
    sigma_n_list.append(sigma_n)
    variance_y_list.append(avg_variance_y)
    variance_pi_B_list.append(avg_variance_pi_B)

# 绘图
plt.figure(figsize=(12, 8))

# 将sigma_n列表转换为浮点数列表并排序
sigma_n_sorted = sorted(sigma_n_list, reverse=True)
variance_y_sorted = [x for _, x in sorted(zip(sigma_n_list, variance_y_list), reverse=True)]
variance_pi_B_sorted = [x for _, x in sorted(zip(sigma_n_list, variance_pi_B_list), reverse=True)]

# 绘制 variance_y 曲线
plt.loglog(sigma_n_sorted, variance_y_sorted, 'b-o', label='variance_y')  # 蓝色线

# 绘制 variance_pi_B_ones_T_g 曲线
plt.loglog(sigma_n_sorted, variance_pi_B_sorted, 'r-o', label='variance_pi_B_ones_T_g')  # 红色线

# 添加图例、标题和坐标轴标签
plt.title('Variance vs sigma_n')
plt.xlabel('sigma_n (log scale)')
plt.ylabel('Variance (log scale)')
plt.legend()

plt.grid(True)  # 添加网格
plt.show()
```
