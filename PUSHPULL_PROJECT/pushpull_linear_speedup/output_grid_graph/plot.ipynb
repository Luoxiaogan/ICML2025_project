```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('./grid, r=4, c=2, n=8, lr=0.03.csv', index_col=0)
df2 = pd.read_csv('./grid, r=4, c=2, n=8, lr=0.04.csv', index_col=0)
df3 = pd.read_csv('./grid, r=4, c=2, n=8, lr=0.05.csv', index_col=0)
df4 = pd.read_csv('./grid, r=4, c=2, n=8, lr=0.06.csv', index_col=0)
df5 = pd.read_csv('./grid, r=4, c=2, n=8, lr=0.07.csv', index_col=0)

import matplotlib.pyplot as plt
plt.semilogy(df1["gradient_norm_on_full_trainset"],label="gradient_norm, lr = 0.03")
plt.semilogy(df2["gradient_norm_on_full_trainset"],label="gradient_norm, lr = 0.04")
plt.semilogy(df3["gradient_norm_on_full_trainset"],label="gradient_norm, lr = 0.05")
plt.semilogy(df4["gradient_norm_on_full_trainset"],label="gradient_norm, lr = 0.06")
plt.semilogy(df5["gradient_norm_on_full_trainset"],label="gradient_norm, lr = 0.07")
plt.title("n=8")
plt.legend()
plt.show()
```
